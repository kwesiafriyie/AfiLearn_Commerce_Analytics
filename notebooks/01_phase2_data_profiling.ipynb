```python
# Initial metadata audits, null check structures, & data dictionary

import os
import sqlite3
import pandas as pd

# Define relative paths to match your folder tree
BASE_DIR = os.path.dirname(os.path.abspath(""))
DB_PATH = os.path.join(BASE_DIR, "data-package", "database", "afilearn_commerce_master.db")

# Open connection
conn = sqlite3.connect(DB_PATH)
cursor = conn.cursor()

# Fetch all active table names to prove the database is populated
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print("=== AFILEARN REPOSITORIES: VERIFIED ACTIVE DB TABLES ===")
for table in cursor.fetchall():
    print(f"✔ Active Table Staging Layer: {table[0]}")
```

```python
target_tables = ["raw_orders", "raw_customers", "raw_order_reviews"]

print("=== MULTI-TABLE DATA INTEGRITY PROFILE ===\n")
for table in target_tables:
    df = pd.read_sql_query(f"SELECT * FROM {table};", conn)
    print(f"📊 PROFILE FOR TABLE: {table} | Total Staged Rows: {df.shape[0]:,}")
    print("-" * 60)
    
    # Calculate null frequencies
    null_counts = df.isna().sum()
    null_percentages = (df.isna().sum() / len(df)) * 100
    
    # Construct an on-the-fly profiling summary table
    profile_df = pd.DataFrame({
        "Data Type": df.dtypes,
        "Missing Rows": null_counts,
        "Null Share (%)": null_percentages.round(2)
    })
    print(profile_df)
    print("\n" + "="*60 + "\n")

# Always close the connection back down
conn.close()
```
